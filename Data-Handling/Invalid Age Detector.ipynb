## Invalid Age Detector

Difficulty: Easyâ€“Medium

Category: Data Validation

Background :

HR kirim data karyawan, tapi ada umur yang tidak masuk akal.
Sebelum dipakai analisis, data ini harus ditandai.

```python
import pandas as pd
import numpy as np

data = {
    "name": ["A", "B", "C", "D", "E"],
    "age": [25, -3, 130, 40, None]
}

df = pd.DataFrame(data)
```

```python
def validate_age(df):
    conditions = [
        (df['age'] >= 0) & (df['age'] <= 100),
        (df['age'].isnull()),
        (df['age'] < 0) | (df['age'] > 120),
    ]
    choices = [True, False, False]
    df['is_valid'] = np.select(conditions, choices, default='unknown')
    return df
    
```

```python
df_clean = validate_age(df)
print(df_clean)
```
