### Dataset

```python
import pandas as pd
import numpy as np

data = {
    "name": ["A", "B", "C", "D", "E", "F"],
    "age": [15, 22, 35, 47, 63, 29]
}

df = pd.DataFrame(data)
```

```python
def add_age_group(df):
    conditions = [
        (df['age'] < 17),
        (df['age'] < 35),
        (df['age'] < 50),
        (df['age'] >= 50)
    ]
    choices = ['Teen', 'Young Adult', 'Adult', 'Senior']
    df['age_group_v2'] = np.select(conditions, choices, default='Unknown')
    return df
```

```python
df_clean = add_age_group(df)
print(df)
```
