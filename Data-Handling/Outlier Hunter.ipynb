## Outlier Hunter

Difficulty: Medium

Category: Outlier Detection

 Background

Tim produk curiga ada data transaksi aneh (terlalu besar) yang bisa ngerusak analisis.

Kamu diminta menandai, bukan menghapus dulu.

```python
import pandas as pd

data = {
    "transaction_id": [1, 2, 3, 4, 5, 6],
    "amount": [100, 120, 130, 125, 5000, 110]
}

df = pd.DataFrame(data)
```

```python
def detect_outlier(df):
    Q1 = df['amount'].quantile(0.25)
    Q3 = df['amount'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df['is_outlier'] = df['amount'].apply(lambda x: x < lower_bound or x > upper_bound)
    return df
```

```python
df_detected = detect_outlier(df)
print(df_detected)
```
