## Sales Summary Bot

Difficulty: Easy–Medium
Category: Data Aggregation

Background :

Manager mau ringkasan cepat:

“Total penjualan tiap kota berapa?”

### Dataset

```python
import pandas as pd

data = {
    "city": ["Jakarta", "Bandung", "Jakarta", "Surabaya", "Bandung", "Jakarta"],
    "sales": [100, 150, 200, 120, 130, 170]
}

df = pd.DataFrame(data)
```

```python
def total_sales_by_city(df):
    return df.groupby('city')['sales'].sum().reset_index()
```

```python
df = total_sales_by_city(df)
print(df)
```
