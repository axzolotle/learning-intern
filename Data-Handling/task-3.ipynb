## Data Handling Intern Level Up

- Difficulty:  Easy–Medium
- Category: Data Cleaning (Duplicates & Standardization)

- Background :

    Tim sales upload data transaksi… tapi beberapa transaksi keinput dua kali dan format teksnya berantakan.

Kamu harus merapikan sebelum masuk database utama.

```python
import pandas as pd

data = {
    "customer_id": [101, 102, 101, 103, 102, 104],
    "product": ["Laptop", "Mouse", "laptop ", "Keyboard", "MOUSE", "Monitor"],
    "price": [1500, 25, 1500, 75, 25, 300]
}

df = pd.DataFrame(data)
```

```python
def clean_transaction_data(df):
    df['product'] = df['product'].str.strip().str.lower()
    df.groupby(['customer_id', 'product', 'price'], as_index=False)
    df = df.drop_duplicates()
    return df
```

```python
df_cleaned = clean_transaction_data(df)
print(df_cleaned)
```
