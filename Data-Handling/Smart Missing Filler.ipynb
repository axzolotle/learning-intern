## Smart Missing Filler

Difficulty:  Medium

Category: Missing Value Handling (Logic-based)

Background :

Tim finance minta data pelanggan, tapi income banyak yang kosong.
Aturannya:

“Kalau income kosong, isi pakai median income dari group pekerjaan yang sama.”

```python
import pandas as pd

data = {
    "name": ["A", "B", "C", "D", "E", "F"],
    "job": ["Engineer", "Engineer", "Designer", "Designer", "Engineer", "Designer"],
    "income": [10_000_000, None, 8_000_000, None, None, 7_500_000]
}

df = pd.DataFrame(data)
```

```python
def fill_income_by_job(df):
    income_mean = df.groupby('job')['income'].mean()
    df['income'] = df.apply(
        lambda row: income_mean[row['job']] if pd.isnull(row['income']) else row['income'],
        axis=1
    )
    return df
```

```python
df_filled = fill_income_by_job(df)
print(df_filled)
```
