## The Missing Intern Report

Difficulty: Easy
Category: Data Cleaning (Pandas)

Background :

Tim marketing ngasih kamu dataset pelanggan, tapiâ€¦ data intern sebelumnya berantakan. Banyak nilai kosong dan format nggak konsisten.

Kamu diminta membersihkan dataset sebelum dipakai analisis.

### Dataset

```python
import pandas as pd

data = {
    "name": ["Andi", "Budi", None, "Citra", "Dewi", None],
    "age": [23, None, 21, 25, None, 22],
    "city": ["Jakarta", "Bandung", "Jakarta", None, "Surabaya", "Bandung"]
}

df = pd.DataFrame(data)
```

```python
def clean_customer_data(df):
    if df.name.isnull().any():
        df = df.dropna(subset=['name'])
    if df.age.isnull().any():
        mean_age = df['age'].mean()
        df['age'] = df['age'].fillna(mean_age)
    if df.city.isnull().any():
        df['city'] = df['city'].fillna('Unknown')
    return df
```

```python
df_cleaned = clean_customer_data(df)
print(df_cleaned)
```
