## Active Customer Filter

Difficulty: Easyâ€“Medium

Category: Boolean Filtering

Background: Tim marketing mau kampanye, tapi hanya ke pelanggan yang benar-benar layak.

```python
import pandas as pd

data = {
    "customer_id": [1, 2, 3, 4, 5],
    "age": [25, 17, 40, 30, 15],
    "is_active": [True, True, False, True, True],
    "income": [5_000_000, 2_000_000, 10_000_000, 4_000_000, 1_000_000]
}

df = pd.DataFrame(data)
```

```python
def filter_eligible_customers(df):
    eligible_customers = df[
        (df['age'] >= 18) &
        (df['is_active'] == True) &
        (df['income'] > 3_000_000)
    ]
    return eligible_customers

```

```python
df_filtered = filter_eligible_customers(df)
print(df_filtered)
```
