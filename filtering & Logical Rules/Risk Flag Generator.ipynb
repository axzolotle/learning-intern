## Risk Flag Generator

Difficulty: Medium
Category: Rule-based Feature

Background :

Tim finance mau menandai pelanggan berisiko untuk dicek manual.

```python
import pandas as pd
import numpy as np

data = {
    "name": ["A", "B", "C", "D", "E"],
    "age": [22, 45, 17, 70, 30],
    "income": [2_500_000, 15_000_000, 1_000_000, 3_000_000, 8_000_000]
}

df = pd.DataFrame(data)
```

```python
def add_risk_flag(df):
    conditoins = [
        (df['age'] < 18) | (df['income'] < 2_000_000),
        (df['age'] >= 65),
        (df['age'] >= 18) & (df['age'] < 65) & (df['income'] >= 2_000_000)
    ]
    choices = ['High', 'Medium', 'Low']
    df['risk_flag'] = np.select(conditoins, choices, default='Unknown')
    return df
```

```python
df_filtered = add_risk_flag(df)    
print(df_filtered)
```
