## Suspicious Record Finder

Difficulty: Mediumâ€“Hard
Category: Edge Case Filtering

Background:

Tim audit mau ngecek data yang secara logika mencurigakan, bukan cuma berdasarkan satu kolom.

```python
import pandas as pd

data = {
    "id": [1, 2, 3, 4, 5],
    "age": [16, 30, 70, 25, 40],
    "employment_status": ["employed", "unemployed", "employed", "employed", "student"],
    "income": [5_000_000, 0, 12_000_000, 1_000_000, 3_000_000]
}

df = pd.DataFrame(data)
```

```python
def find_suspicious_records(df):
    suspicious_records = df[
        ((df['age'] < 18) & (df['employment_status'] == 'employed')) |
        ((df['income'] == 0) & (df['employment_status'] == 'employed')) |
        ((df['age'] > 65) & (df['income'] < 2_000_000))
    ]
    return suspicious_records
```

```python
df_filtered = find_suspicious_records(df)
print(df_filtered)
```
