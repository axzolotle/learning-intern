## Multi-Condition Segment Builder

Difficulty: Medium

Category: Chained Logical Rules

 Background:

Tim bisnis mau segmentasi pelanggan untuk strategi berbeda.
Segment harus eksklusif (satu pelanggan = satu segmen).

```python
import numpy as np
import pandas as pd

data = {
    "customer_id": [1, 2, 3, 4, 5, 6],
    "age": [19, 28, 35, 42, 67, 23],
    "income": [2_000_000, 6_000_000, 3_000_000, 10_000_000, 4_000_000, 1_500_000]
}

df = pd.DataFrame(data)
```

```python
def assign_customer_select(df):
    conditions = [
        (df['age'] < 21) | (df['income'] < 2_000_000),
        (df['age'] >= 21) & (df['income'] >= 8_000_000),
        (df['age'] > 60),
        (df['age'] >= 21) & (df['age'] <= 60) & (df['income'] >= 2_000_000) & (df['income'] < 8_000_000)
    ]
    choices = ['basic', 'premium', 'senior', 'standard']
    df['customer_segment'] = np.select(conditions, choices, default='unknown')
    return df
```

```python
def assign_customer_where(df):
    df['customer_segment'] = np.where(
        (df['age'] < 21) | (df['income'] < 2_000_000), 
        'basic',
        np.where(
            (df['age'] >= 21) & (df['income'] >= 8_000_000),
            'premium',
            np.where(
                df['age'] > 60,
                'senior',
                'standard'
            )
        )
    )
    return df
```

```python
df_filtered = assign_customer_where(df)
print(df_filtered)
```
